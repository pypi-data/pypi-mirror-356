```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, Slider
from bokeh.layouts import column
from bokeh.io import push_notebook
import numpy as np
import time
import threading

output_notebook()
```

```python
# Data source for the plot
source = ColumnDataSource(data=dict(x=[], y=[]))

# Create a Bokeh figure
plot = figure(title="Live Sine Wave", x_axis_type='datetime', width=800, height=300)
plot.line('x', 'y', source=source, line_width=2)

# Add a slider to control frequency
slider = Slider(start=0.1, end=5.0, value=1.0, step=0.1, title="Frequency (Hz)")

# Show the layout (plot + slider)
handle = show(column(slider, plot), notebook_handle=True)

# Shared state
running = True
```

```python
# Background update thread
def update_loop():
    while running:
        print('running')
        t = time.time()
        freq = slider.value
        new_x = [t * 1000]  # bokeh wants ms timestamps
        new_y = [np.sin(2 * np.pi * freq * t)]
        source.stream({'x': new_x, 'y': new_y}, rollover=300)
        print('streamed')
        push_notebook(handle=handle)
        time.sleep(0.1)  # update every 100 ms

# Start the background thread
threading.Thread(target=update_loop, daemon=True).start()
```
